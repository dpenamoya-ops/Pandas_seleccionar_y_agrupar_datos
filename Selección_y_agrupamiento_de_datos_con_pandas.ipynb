### Ajustando la base de datos

Slo nos interesan las emisiones de gases generados por los estados que estn en el inventario nacional, por lo que identificaremos los tipos de datos nicos en la columna *Emisso / Remoo / Bunker*.

---

- *Remoes son datos referentes a la eliminacin de gases de efecto invernadero, que es lo contrario a las emisiones*
- *NCI son los datos no includos en el inventario nacional*
- *Bunkers son emisiones del transporte martimo y areo internacional, por lo tanto no son emisiones directas de los estados.*

*Informacin disponible en el sitio web de SEEG: https://seeg.eco.br/entenda-as-estimativas*

---

```python
emision_gases['Emisso / Remoo / Bunker'].unique()
```

Primero, vamos a revisar los datos de remoo(eliminacin), para identificar si todos ellos son de retirada de gases de efecto invernadero.

```python
(emision_gases['Emisso / Remoo / Bunker'] == 'Remoo NCI') | (emision_gases['Emisso / Remoo / Bunker'] == 'Remoo')
```

```python
emision_gases[emision_gases['Emisso / Remoo / Bunker'].isin(['Remoo NCI', 'Remoo'])]
```

```python
emision_gases.loc[emision_gases['Emisso / Remoo / Bunker'].isin(['Remoo NCI', 'Remoo']), 1970:2021]
```

Para identificar si todos los valores son negativos, vamos a utilizar la funcin `max()` para verificar si existe algun valor mayor que 0.

```python
emision_gases.loc[emision_gases['Emisso / Remoo / Bunker'].isin(['Remoo NCI', 'Remoo']), 1970:2021].max()
```

Vamos tambien a identificar si algun valor de tipo *Bunker* corresponde a alguna emisin hecha por algn estado.

```python
emision_gases.loc[emision_gases['Emisso / Remoo / Bunker'] == 'Bunker', 'Estado'].unique()
```

```python
emision_gases = emision_gases[emision_gases['Emisso / Remoo / Bunker'] == 'Emisso']
emision_gases
```

Como no sobr ningn valor til para analizar en la columna *Emisso / Remoo / Bunker*, vamos a retirarla por completo de la base de datos.

```python
emision_gases = emision_gases.drop(columns = 'Emisso / Remoo / Bunker')
emision_gases
```

#2. Agrupamiento de los datos

```python
emision_gases.loc[:,'Nvel 1 - Setor':'Produto'].columns
```

```python
columnas_info = list(emision_gases.loc[:,'Nvel 1 - Setor':'Produto'].columns)
columnas_info
```

```python
emision_gases.loc[:,1970:2021].columns
```

```python
columnas_emision = list(emision_gases.loc[:,1970:2021].columns)
columnas_emision
```

```python
emision_gases.melt(id_vars = columnas_info, value_vars = columnas_emision, var_name = 'Ao', value_name = 'Emisiones')
```

```python
emisiones_por_ao = emision_gases.melt(id_vars = columnas_info, value_vars = columnas_emision, var_name = 'Ao', value_name = 'Emisiones')
emisiones_por_ao
```

##Analisis de las emisiones

```python
emisiones_por_ao.groupby('Gs')
```

```python
emisiones_por_ao.groupby('Gs').groups
```

```python
emisiones_por_ao.groupby('Gs').get_group('CO2 (t)')
```

```python
emisiones_por_ao.groupby('Gs').sum(numeric_only=True)
```

```python
emisiones_por_gas = emisiones_por_ao.groupby('Gs').sum(numeric_only=True).sort_values('Emisiones', ascending = False)
emisiones_por_gas
```

```python
emisiones_por_gas.plot(kind = 'barh', figsize = (10, 6));
```

```python
emisiones_por_gas.iloc[0:9]
```

```python
float(emisiones_por_gas.iloc[0:9].sum()/emisiones_por_gas.sum())*100
```

```python
print(f'Las emisiones de CO2 coresponden a {float(emisiones_por_gas.iloc[0:9].sum()/emisiones_por_gas.sum())*100:.2f}% de las emisiones totales de gases de efecto invernadero en Brasil de 1970 a 2021')
```
